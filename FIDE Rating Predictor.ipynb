```python
import matplotlib.pyplot as plt
rating_over_a_year = chess[['Player','Jan Rtg', 'Feb Rtg', 'Mar Rtg', 'April Rtg', 'May Rtg','June Rtg','July Rtg','Aug Rtg','Sep Rtg','Oct Rtg',' Nov Rtg','Dec Rtg']].head(15)
numpy = rating_over_a_year.to_numpy()
flat = np.array(numpy).flatten()
#print(flat.size) gives 195 elements
reshape = flat.reshape(15,13) #reshaping into 15*13 matrix
x=["2022-Jan","2022-Feb","2022-March","2022-April","2022-May","2022-June","2022-July","2022-Aug","2022-Sep","2022-Oct","2022-Nov","2022-Dec"]
plt.figure(figsize=(17,8)) 
for i in range (np.size(reshape,0)):
    y=[]
    for j in range (1,np.size(reshape,1)):
        y.append(reshape[i][j])
    plt.title("Ratings of top-15 chess players", fontsize=20)
    plt.style.use('fivethirtyeight')
    plt.xticks(rotation=45) 
    plt.xlabel('Time Span',fontsize=16)
    plt.ylabel('FIDE Rating',fontsize=16)
    plt.plot(x,y,linewidth=3,label=reshape[i][0])
    plt.legend(bbox_to_anchor=(1.05,1) )
    plt.tight_layout()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
chess = pd.read_csv('FIDE_LIST.csv')
X_data1 = chess[['Black', 'Wins(Black)', 'Loss(Black)', 'Draws(Black)']].head(15).to_numpy()
X_data2 = chess[['White', 'Wins(White)', 'Loss(White)', 'Draws(White)']].head(15).to_numpy()
curr_rating = chess['C.Rating'].head(15).to_numpy()
#Wins with black are given more priority
scaling_factor = 1.2
#Loss with white are given more deduction points
k_factor = 15
#Draws with white are given less deduction points
#Draws with black are given less increment points
win_ratio_white = X_data2[:, 1] / X_data2[:, 0]*k_factor*1
draw_ratio_white = X_data2[:, 3] / X_data2[:, 0]*k_factor*-1.1
loss_ratio_white = X_data2[:, 2] / X_data2[:, 0]*k_factor*-scaling_factor
win_ratio_black = X_data1[:, 1] / X_data1[:, 0]*k_factor*scaling_factor
draw_ratio_black = X_data1[:, 3] / X_data1[:, 0]*k_factor*+1.1
loss_ratio_black = X_data1[:, 2] / X_data1[:, 0]*k_factor*-1
total_rating_change = win_ratio_white + draw_ratio_white + loss_ratio_white + win_ratio_black + draw_ratio_black + loss_ratio_black
X_test = np.column_stack((X_data1, X_data2))
y = total_rating_change
X_train, X_val, y_train, y_val = train_test_split(X_test, y, test_size=0.2, random_state=42)
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_val)
mse = mean_squared_error(y_val, y_pred)
print("Mean Squared Error:", mse)
new_rating_change = model.predict(X_test)
print("Calculated New Rating Change:", total_rating_change)
print("Predicted New Rating Change:", new_rating_change)
```

```python
# Plot the predicted rating changes against the actual rating changes
plt.figure(figsize=(10, 6))
plt.scatter(total_rating_change,new_rating_change, color='blue')
plt.plot(total_rating_change,total_rating_change,color='red', linestyle='--')
plt.title('Predicted vs. Actual Rating Changes')
plt.xlabel('Actual Rating Changes')
plt.ylabel('Predicted Rating Changes')
plt.grid(True)
plt.show()
```

```python
calculated_final_rating=curr_rating+total_rating_change
predicted_final_rating=curr_rating+new_rating_change
chess['Total Rating Change'] = total_rating_change
chess['New Rating Change'] = new_rating_change
chess['Calculated Final Rating'] = calculated_final_rating
chess['Predicted Final Rating'] = predicted_final_rating
chess.to_csv('FIDE_LIST.csv')
chess
```
